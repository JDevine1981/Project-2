```python
import pandas as pd
```

```python
# Load in Raw Data
title_basics = pd.read_csv('Data/movie_basics.csv')
title_basics.info()
title_basics.head(5)
```

```python
title_basics = title_basics.drop(columns='Unnamed: 0')
title_basics
```

```python
title_basics['genres'].value_counts()
```

```python
## create a col with a list of genres
title_basics['genres_split'] = title_basics['genres'].str.split(',')
title_basics
```

```python
exploded_genres = title_basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = title_basics['genres_split'].explode().unique()
unique_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
title_genres = exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
title_basics.head()
```

```python
genre_lookup = pd.DataFrame({'Genre_Name': genre_map.keys(), 'Genre_ID': genre_map.values()})
genre_lookup.head()
```
